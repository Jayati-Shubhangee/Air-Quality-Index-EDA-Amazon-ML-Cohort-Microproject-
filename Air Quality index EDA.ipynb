```python
# Step 1 data loading part
df = pd.read_csv(r"C:\Users\jayat\Downloads\AIR QUALITY INDEX DATASET\city_day.csv")
df.head()
```

```python
# STEP 2 Basic Info & Structure (command for data frame info) 
df.shape
df.info()
df.describe()
```

```python
# STEP -3 Missing Value Handling (command for checking missing values)
df.isnull().sum()
```

```python
# STEP 4 cleaning the data for analysis 
#keeping most relevant columns 
# drop rows that have missing AQI or Major Pollutant values
# Keep only important columns for now
df = df[['City', 'Date', 'PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2', 'AQI', 'AQI_Bucket']]

# Drop rows where AQI is missing (important for insights)
df = df.dropna(subset=['AQI'])

# Fill missing values for pollutants using column means (safer than drop)
for col in ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
#again checking for null values 
df.isnull().sum()
```

```python
# STEP 5 Univariate Analysis EDA Visulaization + Insights 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['AQI'], bins=50, kde=True, color='skyblue')
plt.title('AQI Distribution Across All Cities')
plt.xlabel('Air Quality Index')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#top 10 most polluted cities by average AQI 
top_cities = df.groupby('City')['AQI'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='Reds_r')
plt.title('Top 10 Most Polluted Cities (Average AQI)')
plt.xlabel('Average AQI')
plt.ylabel('City')
plt.show()
```

```python
#AQI overall trend 
df['Date'] = pd.to_datetime(df['Date'])
df_grouped = df.groupby('Date')['AQI'].mean()

plt.figure(figsize=(12, 5))
df_grouped.plot()
plt.title('Overall AQI Trend Over Time')
plt.ylabel('Average AQI')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

```python
#Correlation Heatmap 
plt.figure(figsize=(10, 7))
sns.heatmap(df[['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2', 'AQI']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Pollutants and AQI')
plt.show()
```

```python
#AQI Category Distribution 
plt.figure(figsize=(7, 5))
sns.countplot(x='AQI_Bucket', data=df, order=df['AQI_Bucket'].value_counts().index, palette='viridis')
plt.title('AQI Category Distribution (Bucket-wise)')
plt.xticks(rotation=45)
plt.ylabel('Number of Days')
plt.show()
```
